The analyzed data was gotten from @WeRateDogs, a twitter account that rates dog photos with humorous tweets. What is different about this account is it's unique style of rating. Unlike many conventional ratings, it is the norm for this account to give a rating score that is higher than the denominator value (10). It is right to say that, as far as @WerateDogs is concerned, every dog is more than perfect.

Since a lot of twitter accounts send in their Dog's pictures to be rated, getting a dog rated is highly competitive. From the bar Chart below, it shows that @WerateDog tweets mostly between 3pm and 6am and no record of tweets between 7am and 1pm on any day.



From the bar chart below, the followers of @WerateDogs are seen to interact best with content posted at 6am

To get the best rated breed, the number of time that the breed was rated was taken into consideration. The breeds considered were breeds that were rated more than 10 times. With this criterion, Samoyed is the best rated breed.

According to the chart, Bedlington Terrier is the breed that drives the most engagement(Retweets + favorites) while Brabracon Griffon drives the least engagement.
